```python
#Reading file usng pandas library
import pandas as pd
url='https://docs.google.com/spreadsheets/d/1cnPUhCbUm5p3DMWw6wcSu4li7qChBrhd-UyRkW2RSUg/edit?usp=sharing'
#Assign the table as database to pandas data frame
mydb=pd.read_html(url)[0]

#Data cleaning and Wrangling

mydb.drop(1) 

mydb.dropna(axis=1, how='all',inplace=True)#Removing NaN Columns
mydb

mydb.dropna(axis=0,how='all',inplace=True) #Removing NaN row
mydb

mydb =mydb.loc[:, ~mydb.columns.str.contains('^Unnamed')]#Removing unnamed column
mydb

mydb.index = range(100)#Changing Index
mydb
df=pd.DataFrame(mydb)# Changing to dataframe from html
df
```

```python
#Creating Engine
import pymysql
from sqlalchemy import create_engine
user="root"
passw="password"
schema="Mydukan"
host="localhost"
port="3306"

engine= create_engine('mysql+pymysql://'+user+':'+ passw+'@'+host+':'+ port+'/'+schema, echo=False)
#converting dataframe to mysql
df.to_sql(name= 'Beginner', con=engine, if_exists = 'replace', index=True)
```
